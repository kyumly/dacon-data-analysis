```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier

data = pd.read_csv('../data/Personal Loan.csv')


data
```

```python


train_data = data.iloc[:, [1,2,3,5,6,7,8,10,11,12,13]]
target_date = data.iloc[:, 9]
```

```python
train_data.describe()
```

```python
import sklearn.model_selection  as pre

train_data, train_test, target_data, target_test = pre.train_test_split(train_data, target_date, random_state=42, test_size=0.3)

print(train_data.shape, train_test.shape, target_data.shape, target_test.shape)
```

```python
data_test = data.iloc[:, 9].to_numpy()
set(data_test)
```

```python
dt = DecisionTreeClassifier(random_state=42)

dt.fit(train_data, target_data)
print(dt.score(train_test, target_test))
```

```python
import matplotlib.pyplot  as plt
from sklearn.tree import plot_tree
plt.figure(figsize=(50, 50))
plot_tree(dt)
plt.show()
```

```python
columns = data.columns.to_numpy()
columns
```

```python
columns=columns.reshape(-1, 1)
columns.shape
print(columns)
```

```python
columns=columns[[1,2,3,5,6,7,8,10,11,12,13]]
```

```python
plt.figure(figsize=(50, 50))
plot_tree(dt, max_depth=3, feature_names=columns)
plt.show()
```
